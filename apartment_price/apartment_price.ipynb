### Обработка пропущенных значений

Чтобы узнать количество пропусков в каждом столбце применю метод isna() в сочетании с методом sum().

```python
sell_apartments.isna().sum()
```

Анализируя столбцы, можно сделать **вывод, что не все пропуски можно заполнить, так как нет подходящих значений на замену.** В таких случаях я оставлю пропуски пустыми. Данный вывод применим к следующим столбцам:
- `ceiling_height`;
- `floors_total`; 
- `airports_nearest`;         
- `city_centers_nearest`;      
- `parks_around_3000`;

Для данных с высотой потолков `ceiling_height` нет подходящих значений на замену. Пропуски в этой колонке могли возникнуть по вине собственника: данные пропущенны намеренно, либо случайно. Тут уже не узнаешь, поэтому оставлю пропущенные значения как есть.

Столбец `floors_total` содержит 86 пропущенных значений. Пропуски в этой колонке могут быть, если на продажу выставлена не квартира, а частное жилье, например, загородный дом. Либо собстенник просто не посчитал нужным указывать общее количество этажей в доме. Поэтому, строки с пропущенными значениями я не буду заменять или удалять из датафрейма.

Столбцы с информацией о расстоянии до центра города `city_centers_nearest`, аэропорта `airports_nearest`, ближайшего парка `parks_nearest` и водоёма `ponds_nearest`, а также число водоемов `ponds_around_3000` и парков `parks_around_3000` в радиусе 3 км получены автоматически на основе картографических данных. Пропуски в этих колонках могли возникнуть из-за недостатка информации для рассчета этих данных, поэтому такие пропуски не нужно заменять характерными значениями и удалять из датафрейма.

Столбец `days_exposition` может содержать пропущенные значения в том случае, если объявление было снято с публикации ровно в тот же день, что и размещено (возможно, покупатель сразу нашелся или собственник передумал и решил снять объявление о продаже квартиры). 

**Методики борьбы с пропусками в других столбцах далее:**

- столбeц `is_apartment` содержит булев тип данных (категориальная переменная): `False` означает, что квартира не апартаменты; `True` - апартаменты. Такое большое количество пропущенных значений в данной колонке связано с тем, что собстенники квартиры оставили этот вопрос пропущенным, подразумевая, что квартира не является апартаментами. Поэтому пропущенные значения я заменю на `False`. Перед этим я проверила уникальные значения в столбце методом unique(), чтобы убедиться, что значение `True` есть в этой колонке.

```python
sell_apartments['is_apartment'].unique()
```

```python
sell_apartments['is_apartment'] = sell_apartments['is_apartment'].fillna(False)
```

- столбец `balcony` содержит 11519 пропущенных значений. Можно сказать, что данные о количестве балконов в этом стобце пропущены, так как в квартире нет балконов. Поэтому заменю пропуски на '0'.

```python
sell_apartments['balcony'] = sell_apartments['balcony'].fillna(0)
```

- в столбце `locality_name` содержится 49 пропущенных значений. Это значит, что строки с пропусками населенного пункта потеряли смысл. Таких строк 0,2% от общих данных датафрейма, поэтому я удалю строки таблицы с пропущенными значениями с помощью метода dropna().

```python
sell_apartments = sell_apartments.dropna(subset=['locality_name']).reset_index(drop=True)
```

- данные в столбцах `living_area` и `kitchen_area` - количественые, и для того чтобы заполнить пропущенные значения в этих столбцах я рассчитаю медианное значение для каждой группы по общей площади квартиры. Чтобы сформировать диапазоны значений в группах, буду использовать метод qcut().

```python
pd.qcut(sell_apartments['total_area'], q=4, precision=0).head()
```

Выделю **4 группы по общей площади квартиры**, которые добавлю в новый столбец `area_category`:
- `маленькая` - площадь в диапазоне `(11.0, 40.0]`;
- `средняя` - площадь в диапазоне `(40.0, 52.0]`;
- `выше среднего` - площадь в диапазоне `(52.0, 70.0]`;
- `большая` - площадь в диапазоне `(70.0, 900.0]`.

```python
labels_area = ['маленькая', 'средняя', 'выше среднего', 'большая']
sell_apartments['area_category'] = pd.qcut(sell_apartments['total_area'], q=4, labels=labels_area, precision=0)
sell_apartments.head()
```

Рассчитаю медиану в каждой группе для заполнения пропущенных значений в столбце `kitchen_area`.

```python
kitchen_area_median = (
    sell_apartments.groupby('area_category')['kitchen_area'].median()
)
kitchen_area_median
```

С помощью метода fillna() заполню пропущенные значения в столбце `kitchen_area` медианой для каждой группы:

```python
sell_apartments['kitchen_area'] = sell_apartments['kitchen_area'].fillna(
    sell_apartments.groupby('area_category')['kitchen_area'].transform('median')
)
```

Рассчитаю медиану в каждой группе для заполнения пропущенных значений в столбце `living_area`.

```python
living_area_median = (
    sell_apartments.groupby('area_category')['living_area'].median()
)
living_area_median
```

С помощью метода fillna() заполню пропущенные значения в столбце `living_area` медианой для каждой группы:

```python
sell_apartments['living_area'] = sell_apartments['living_area'].fillna(
    sell_apartments.groupby('area_category')['living_area'].transform('median')
)
```

Избавлюсь от е-формата чисел в столбце `last_price`, т.е. приведу к единицам измерения млн.руб:

```python
sell_apartments['last_price'] = sell_apartments['last_price'] / 1000000
```

### Проверка на неявные дубликаты

Методом unique() проверю столбец `locality_name` с категориальными переменными на неявные дубликаты:

```python
sell_apartments['locality_name'].unique()
```

Чтобы избавиться от неявных дубликатов, уберу из названия населенного пункта его принадлежность.

```python
excluded_words = {
    'посёлок', 'поселок', 'деревня', 'городского',
    'типа', 'село', 'коттеджный', 'садовое',
    'товарищество', 'городской', 'при', 'железнодорожной',
    'станции', 'садоводческое', 'некоммерческое'
}
```

Напишу функцию, которая на вход принимает строку с названием населенного пункта. На выходе функция возвращает новое название с исключенными словами из множества `excluded_words`.

```python
def replace_words(value):
    words = value.split()
    new_locality_name = []
    for word in words:
        if word not in excluded_words:
            new_locality_name.append(word)
    return ' '.join(new_locality_name)

sell_apartments['locality_name'] = sell_apartments['locality_name'].apply(replace_words)
```

Снова проверяю уникальные значения: после преобразований названий населенных пунктов стало 305, до преобразований их было 364.

```python
sell_apartments['locality_name'].unique()
```

**Вывод:** от неявных дубликатов в категориальном столбце удалось избавиться и сократить название населенных пунктов на 59 значений.

## Расчёты и добавление результатов в таблицу

Для дальнейшего анализа данных рассчитаю ряд значений и добавлю их в таблицу:

- *цена квадратного метра:*

```python
sell_apartments['price_square'] = sell_apartments['last_price'] * 1000000 / sell_apartments['total_area']
sell_apartments['price_square'] = sell_apartments['price_square'].astype('int')
```

- *день недели, месяц и год публикации объявления:*

```python
sell_apartments['weekday_publication'] = sell_apartments['date_publication'].dt.weekday
```

```python
sell_apartments['month_publication'] = sell_apartments['date_publication'].dt.month
```

```python
sell_apartments['year_publication'] = sell_apartments['date_publication'].dt.year
```

- *этаж квартиры*, сгруппированный по трем категориям: первый, последний, другой:

```python
def group_floors(row):
    if row['floor'] == 1:
        return 0
    elif not pd.isna(row['floors_total']) and row['floor'] == row['floors_total']:
        return 1
    else:
        return 2
    
sell_apartments['floor_group'] = sell_apartments.apply(group_floors, axis=1)
```

Вынесу в отдельный датафрейм `floor_group` информацию о категориях этажей и их наименованиях:

```python
data = {'floor_group': [0, 1, 2], 'name': ['первый', 'последний', 'другой']}
floor_group = pd.DataFrame.from_dict(data)
floor_group
```

- *отношение площади кухни к общей площади квартиры:*
